```python
import random
import time

def play_game():
    score = 0
    rounds = 5

    print("⚔️ Welcome to Number Ninja!")
    print("Guess the secret number before time runs out!\n")

    for r in range(1, rounds + 1):
        number = random.randint(1, 50)
        attempts = 0
        max_time = 10

        print(f"\n🎯 Round {r}")
        print("Guess a number between 1 and 50")
        start_time = time.time()

        while True:
            if time.time() - start_time > max_time:
                print("⏳ Time's up!")
                print(f"The number was: {number}")
                break

            guess = input("Your guess: ")

            if not guess.isdigit():
                print("❌ Enter a valid number!")
                continue

            guess = int(guess)
            attempts += 1

            if guess == number:
                print("🎉 Correct!")
                gained = max(10 - attempts, 1)
                score += gained
                print(f"⭐ You earned {gained} points!")
                break
            else:
                diff = abs(guess - number)
                if diff <= 3:
                    print("🔥 VERY HOT!")
                elif diff <= 7:
                    print("🌡️ Warm")
                else:
                    print("❄️ Cold")

    print("\n🏁 Game Over!")
    print(f"Final Score: {score}")

play_game()
```
